```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# STEP-2 Input Data
x = np.array([1,2,3]).reshape(-1,1)
y = np.array([2,3,5])
```

```python
#step-3 create model and fit data
model = LinearRegression()
model.fit(x,y)
```

```python
#step-4 get coefficient and print
w = model.coef_[0]
b = model.intercept_
print(f"The slope is: {w} ")
print(f"The intercept is: {b} ")
```

```python
#step-5 make predictions
y_pred = model.predict(x)
print("\nprediction for training data")
for xi,yi,ypi in zip(x.flatten(), y, y_pred):
    print(f"x = {xi}, Actual y ={yi}, Predicted y = {ypi}")
```

```python
#step-6: calculate error
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared error is:{mse}")
print(f"R2_Score is: {r2}")
```

### LR using MLE

```python
#step1 imports
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy.optimize import minimize
```

```python
# STEP-2 Input Data
x = np.array([1,2,3]).reshape(-1,1)
y = np.array([2,3,5])
```

```python
# step-3 Negative log Likelyhood function
def neg_log_likelihood(params):
    w, b = params
    sigma2= 1 # asume variance =1
    y_pred = w*x + b
    nll = 0.5*np.sum((y-y_pred)**2 / sigma2)
    return nll
```

```python
# initial values
initial_guess= [0,0]
```

```python
# step-4 minimize nll
result = minimize(neg_log_likelihood, initial_guess)
w_mle, b_mle = result.x
print (f"Slope is: {w_mle}")
print (f"Intercept is: {b_mle}")
```

```python
#step-5 prediction
y_pred = w_mle * x + b_mle
print("\nprediction for training data")
for xi,yi,ypi in zip(x.flatten(), y, y_pred):
    print(f"x = {xi}, Actual y ={yi}, Predicted y = {ypi}")
```

```python
#step-6: calculate error
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared error is:{mse}")
print(f"R2_Score is: {r2}")
```

### LR using GD

```python
#step1 imports
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# STEP-2 Input Data
x = np.array([1,2,3])
y = np.array([2,3,5])
```

```python
#step-3 
w, b = 0,0
alpha = 0.001
n_iter = 1000
n = len(x)
```

```python
for i in range(n_iter):
    y_pred = w*x.flatten() + b
    dw = (-2/n)*np.sum(x.flatten()*(y-y_pred))
    db = (-2/n)*np.sum(y-y_pred)
    w = w - alpha * dw
    b -= alpha*db
print(f"Slope is: {w}")
print(f"Bias is: {b}")
```

### LR using GD with single parameter

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# STEP-2 Input Data
x = np.array([1,2,3])
y = np.array([2,3,5])
n = len(x)
```

```python
#step-3 Calculate loss Function
def loss(w1):
    w0 = np.mean(y) - w1*np.mean(x)
    y_pred = w1*x + w0
    return np.sum((y-y_pred)**2)
```

```python
#step-4 calculate gradient of j wrt w1
def gradient(w1):
    w0 = np.mean(y) - w1*np.mean(x)
    y_pred = w1*x + w0
    return -2*np.sum(x*(y-y_pred))
```

```python
#step5 gradient decent
lr = 0.1
w1 = 4
iter = 15
w1_values = []
loss_values = []

for i in range(iter):
    w1_values.append(w1)
    loss_values.append(loss(w1))
    grad = gradient(w1)
    w1 -= lr* grad
```

```python
#step-6  plot loss function and GD
w_space = np.linspace(-2,4,200)
loss_space = [loss(w) for w in w_space]

plt.figure(figsize = (9,6))
plt.plot(w_space, loss_space, label="J (w1)")
plt.scatter(w1_values, loss_values,color="red" , label="GD steps")
plt.xlabel("w1 (Slope)")
plt.ylabel("Loss (J)")
plt.title("Gradient descent using Slope")
plt.legend()
plt.grid(True)
plt.show()
```

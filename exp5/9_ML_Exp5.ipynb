```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import numpy as np
import pandas as pd
```

```python
# Load real dataset
data = load_breast_cancer()

X = data.data
y = data.target

print("Classes: ", data.target_names)
# malignant, benign
```

```python
df = pd.DataFrame(X)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Logistic Regression

lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)
print("Logistic TestAccuracy: ",accuracy_score(y_test, pred_lr))
pred_lr_train = lr.predict(X_train)
print("Logistic TrainAccuracy: ",accuracy_score(y_train, pred_lr_train))
```

```python
# Decision Tree

dt =  DecisionTreeClassifier()
dt.fit(X_train, y_train)
pred_dt = dt.predict(X_test)
print("Decision TestAccuracy: ",accuracy_score(y_test, pred_dt))
pred_dt_train = dt.predict(X_train)
print("Decision TrainAccuracy: ",accuracy_score(y_train, pred_dt_train))
```

```python
# KNN

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
print("KNN TestAccuracy: ",accuracy_score(y_test, pred_knn))
pred_knn_train = knn.predict(X_train)
print("KNN TrainAccuracy: ",accuracy_score(y_train, pred_knn_train))
```

```python
# Naive Bayes

nb = GaussianNB()
nb.fit(X_train, y_train)
pred_nb = nb.predict(X_test)
print("Naive Bayes TestAccuracy: ",accuracy_score(y_test, pred_nb))
pred_nb_train = nb.predict(X_train)
print("Naive Bayes TrainAccuracy: ",accuracy_score(y_train, pred_nb_train))
```

```python
# MLP Neural Network

mlp = MLPClassifier()
mlp.fit(X_train, y_train)
pred_mlp = mlp.predict(X_test)
print("MLP (NN) TestAccuracy: ",accuracy_score(y_test, pred_mlp))
pred_mlp_train = mlp.predict(X_train)
print("MLP (NN) TrainAccuracy: ",accuracy_score(y_train, pred_mlp_train))
```

```python
# Example confusion matrix, Type-I and Type-II error for LR

cm_lr = confusion_matrix(y_test, pred_lr)
print("\nLinear Regression Confusion Matrix")
print(cm_lr)
TN1, FP1, FN1, TP1 = cm_lr.ravel()

type1_lr = FP1 / (FP1 + TN1) # False positive rate
type2_lr = FN1 / (FN1 + TP1) # False negative rate

print("FP (Type-I): ",FP1)
print("FN (Type-II): ",FN1)
print("Type-I error rate in %:", type1_lr*100)
print("Type-II error rate in %:", type2_lr*100)

```

```python
# Example confusion matrix, Type-I and Type-II error for DT

cm_dt = confusion_matrix(y_test, pred_dt)
print("\Decision Tree Confusion Matrix")
print(cm_dt)
TN2, FP2, FN2, TP2 = cm_dt.ravel()

type1_dt = FP2 / (FP2 + TN2) # False positive rate
type2_dt = FN2 / (FN2 + TP2) # False negative rate

print("FP (Type-I): ",FP2)
print("FN (Type-II): ",FN2)
print("Type-I error rate in %:", type1_dt*100)
print("Type-II error rate in %:", type2_dt*100)
```

```python
# Example confusion matrix, Type-I and Type-II error for NB

cm_nb = confusion_matrix(y_test, pred_nb)
print("\nNaive Bayes Confusion Matrix")
print(cm_nb)
TN4, FP4, FN4, TP4 = cm_nb.ravel()

type1_nb = FP4 / (FP4 + TN4) # False positive rate
type2_nb = FN4 / (FN4 + TP4) # False negative rate

print("FP (Type-I): ",FP4)
print("FN (Type-II): ",FN4)
print("Type-I error rate in %:", type1_nb*100)
print("Type-II error rate in %:", type2_nb*100)
```

```python
# Calculating values for LR

precision = precision_score(y_test, pred_lr)
recall = recall_score(y_test, pred_lr)
f1 = f1_score(y_test, pred_lr)
roc = roc_auc_score(y_test, pred_lr)

print("precision:", precision)
print("recall:", recall)
print("f1:", f1)
print("roc:", roc)

```
